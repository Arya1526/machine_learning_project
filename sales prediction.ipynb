```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.impute import SimpleImputer

from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression

from sklearn.metrics import mean_squared_error, r2_score
```

```python
import pandas as pd
file_path = "/Users/Aryaanil/Documents/Advertising Budget and Sales.csv"
df = pd.read_csv(file_path)
print(df)
```

```python
num_cols = df.select_dtypes(include=np.number).columns
cat_cols = df.select_dtypes(exclude=np.number).columns


num_imputer = SimpleImputer(strategy='mean')
df[num_cols] = num_imputer.fit_transform(df[num_cols])


if len(cat_cols) > 0:
    cat_imputer = SimpleImputer(strategy='most_frequent')
    df[cat_cols] = cat_imputer.fit_transform(df[cat_cols])
```

```python
if len(cat_cols) > 0:
    le = LabelEncoder()

    for col in cat_cols:
        df[col] = le.fit_transform(df[col])
```

```python
if 'TV' in df.columns and 'Radio' in df.columns:
    df['TV_Radio_Total'] = df['TV'] + df['Radio']

if 'TV' in df.columns and 'Newspaper' in df.columns:
    df['TV_Newspaper_Total'] = df['TV'] + df['Newspaper']
```

```python
X = df.drop('Sales ($)', axis=1)
y = df['Sales ($)']
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled,
    y,
    test_size=0.2,
    random_state=42
)

print("X_train shape :", X_train.shape)
print("X_test shape  :", X_test.shape)
print("y_train shape :", y_train.shape)
print("y_test shape  :", y_test.shape)
```

```python
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

lr_pred = lr_model.predict(X_test)
```

```python
rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(X_train, y_train)

rf_pred = rf_model.predict(X_test)
```

```python
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_r2 = r2_score(y_test, lr_pred)

rf_rmse = np.sqrt(mean_squared_error(y_test, rf_pred))
rf_r2 = r2_score(y_test, rf_pred)
```

```python
print("\n===== Linear Regression =====")
print("RMSE :", lr_rmse)
print("R2 Score :", lr_r2)

print("\n===== Random Forest Regressor =====")
print("RMSE :", rf_rmse)
print("R2 Score :", rf_r2)
```

```python
if rf_r2 > lr_r2:
    print("\nRandom Forest Regressor performed better.")
else:
    print("\nLinear Regression performed better.")
```
